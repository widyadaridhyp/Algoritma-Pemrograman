```python
from matplotlib import pyplot as n
```

```python
x=[]#Nilai awalnya himpunan kosong karena kertasnya masih kosong
y=[]

n.xlim(-5,5)#Ini untuk interval kertas. Bebas.
n.ylim(-5,5)

print('Selamat datang di program corat-coret')#Dari sini sampai print('4.....'), untuk opening
print('Ini ada papan corat-coret masih kosong')

n.show()#Untuk menampilkan kertasnya yang sudah memiliki interval yang tadi dituliskan

print('Isi dengan titik koordinat ya')
print('1. Untuk tambahkan titik')
print('2. Untuk lihat hasil coretan')
print('3. Untuk hapus semua coretan')
print('4. Untuk menghentikan program')

while True:
    opsi=int(input('Pilih: '))#Untuk memasukkan opsi
    if opsi==1:#Digunakan untuk membuat coretan atau menambahkan coretan pada kertas
        x.append(eval(input('Masukkan titik-titik x: ')))#Nilai x yang tadinya himpunan kosong akan ditambahkan ini. Nilai yang dimasukkin nanti sebagai titik x yang kita ingin coret. Jika banyak titiknya akan membentuk garis. Jika nilainya lebih dari 1 gunakan tanda koma. 
        y.append(eval(input('Masukkan titik-titik y: ')))#Nilai y yang tadinya himpunan kosong akan ditambahkan ini. Nilai yang dimasukkin nanti sebagai titik x yang kita ingin coret. Jika banyak titiknya akan membentuk garis. Jika nilainya lebih dari 1 gunakan tanda koma. 
        
    if opsi==2:#Opsi 2 digunakan untuk menampilkan coretan pada kertas
        if x==[] and y==[]:
            n.xlim(-5,5)#Kertasnya dimasukkin lagi syntaxnya agar setelah dihapus dengan menggunakan opsi 3 akan menampilkan kertas kosong.
            n.ylim(-5,5)
        for i in range(len(x)):#Untuk menentukan plotnya
            n.plot(x[i],y[i])
        n.show()
        
    if opsi==3:#Opsi 3 digunakan untuk menghapus coretannya
        x=[]
        y=[]
        
    if opsi==4: #Opsi 4 digunakan untuk mengakhiri permainan
        print ('Terima kasih telah bermain!')
        break
```

Program ini adalah program untuk mencoret-coret kertas. Idenya adalah himpunan kosong akan ditambahkan hingga membentuk garis atau coretan. Terdapat 4 opsi, yaitu untuk membuat coretan, memperlihatkan hasil coretan , menghapus coretan, dan memberhentikan program.
Algoritma:
1. Membuat list kosong
2. Membuat kertas dengan interval tertentu
3. Print kata-kata untuk opening
4. Memasukkan opsi yang akan kita pilih
5. Pada opsi 1, akan ditambahkan titik-titik ke list kosong sebelumnya.
6. Pada opsi 2, akan ditampilkan hasil coretannya.
7. Pada opsi 3, coretan tadi akan dihapus
8. Pada opsi 4, permainan akan diberhentikan
